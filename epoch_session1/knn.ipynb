```python
#TASK: to make your own knn model from scratch using only numpy library

import numpy as np

data = [
    [150, 7.0, 1, 'Apple'],
    [120, 6.5, 0, 'Banana'],
    [180, 7.5, 2, 'Orange'],
    [155, 7.2, 1, 'Apple'],
    [110, 6.0, 0, 'Banana'],
    [190, 7.8, 2, 'Orange'],
    [145, 7.1, 1, 'Apple'],
    [115, 6.3, 0, 'Banana']
]  # weight in g,size in cm,color 0:yellow 1:red 2:orange,fruit name

```

```python
#dividing data into feature matrix(X) and label array(y)

X=np.array(data)[:,0:3].astype(float) #1st one is for number of rows and 2nd one is for number of columns. coluns we need only first 3(features).......
y=np.array(data)[:,-1]#same logic. all rows needed but only last column(fruitname) required

print(X,"\n",y,"\n","\n")
print(X.shape,"\n",y.shape)
#NOTE: astype(float) is done because by default the data also has apple and banana so it is string type. this gives error in euclidean function defined later
```

```python
#euclidean distance function for nd space
def euclidean_distance(x1,x2):#x1 and x2 are arrays of same shape
    return np.sqrt(np.sum((x1-x2)**2))


print(euclidean_distance(X[0],X[1]))
```

```python
#step 3: creating a knn class

class KNN:
    
    def __init__(self, k=3):
        self.k = k          # putting value of k
    
    def fit(self, X, y):
        self.X_train = X    # store the training data features
        self.y_train = y    # store the labels
    
    def predict(self, X_test):
        predictions = []
        for x in X_test:                        # loop over each test point
            predictions.append(self.predict_one(x))
        return np.array(predictions)
    

    #NOTE: in python we can use a function before defining it. but we cant call it in runtime before defining it
    
    def predict_one(self, x):
        # step 1: calculate distance from x to every training point
        distances = []
        for x_train in self.X_train:
            d = euclidean_distance(x, x_train)
            distances.append(d)
        distances = np.array(distances)
        
        # step 2: find the k nearest neighbors (get their indices, sorted by distance)
        k_indices = np.argsort(distances)[:self.k]
        
        # step 3: get the labels of those k neighbors
        k_labels = self.y_train[k_indices]
        
        # step 4: vote — return the most common label
        unique_labels, counts = np.unique(k_labels, return_counts=True)
        return unique_labels[np.argmax(counts)]
```

```python
test_data = np.array([
    [118, 6.2, 0],  # Expected: Banana
    [160, 7.3, 1],  # Expected: Apple
    [185, 7.7, 2]   # Expected: Orange
])

knn = KNN(k=3)
knn.fit(X, y)

predictions = knn.predict(test_data)
print(predictions)
```
